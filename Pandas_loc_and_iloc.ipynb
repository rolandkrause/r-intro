# Pandas: Location access and slice

The curated COVID-19/Luxembourg data is at https://github.com/MADS-PY-R-2021/datasets/blob/main/luxembourg_covid19_data.csv

## Create a DataFrame

```python
# importing pandas and numpy
import pandas as pd
import numpy as np

# crete a sample dataframe
data = pd.read_csv("https://raw.githubusercontent.com/MADS-PY-R-2021/datasets/main/luxembourg_covid19_data.csv")
# view the data
data
```

```python
data.set_index('Dated', inplace=True)
data.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
data
```

# loc

```python
data.loc['2021-09-26']
```

```python
data.loc['2021-09-26','Normal care']
```

```python
data.loc[['2021-09-26','2021-09-27']]
```

```python
data.loc[['2021-09-26','2021-09-27'],['Normal care','Intensive care']]
```

```python
data.loc[:,['Normal care','Intensive care']]
```

## Find all the rows based on any condition in a column

```python
# select all rows with a condition
data.loc[data['Intensive care'] >= 5]
```

## Find all the rows with more than one condition

```python
# select with multiple conditions
data.loc[(data['Intensive care'] >= 5) & (data['Death per day'] > 0)]
```

## Select a range of rows using loc

```python
#slice
data.loc['2021-09-03':'2021-09-21']
```

## Select only required columns with a condition

```python
# select few columns with a condition
data.loc[data['Intensive care'] >= 5, ['Normal care','Intensive care']]
```

## Update the values of a particular column on selected rows

```python
# update a column with condition
data.loc[(data['Intensive care'] >= 5), ['Hospital_load']] = 'High'
data.loc[(data['Intensive care'] < 5), ['Hospital_load']] = 'Low'
data
```

# iloc

## Select rows with indices using iloc

```python
# select rows with indexes
data.iloc[[30,60,90,120]]
```

## Select rows with particular indices and particular columns

```python
# select rows with particular indexes and particular columns
data.iloc[[30,60,90,120],[0,1,-3,-1]]
```

## Select a range of rows using iloc

```python
# select a range of rows
data.iloc[365:395]
```

## Select a range of rows and columns using iloc

```python
# select a range of rows and columns
data.iloc[365:395,-4:]
```
